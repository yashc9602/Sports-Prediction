```python
import ujson as json

print("Parsing JSON dataset (this may take a while)")
with open('/kaggle/input/tokyo-2021-olympics-complete-grouped-by-type/tokyo-olympics-all-by-type-2021-08-11.json', 'r') as file:
    olympicsData = json.load(file)
    
countries = olympicsData["Organisation"]
finalData = []

for itemId, item in countries.items():
    item = item["attributes"]
    finalData.append({
        "name": item["name"],
        "id": item["externalId"],
        "goldCount": item["statistics"]["gold"],
        "silverCount": item["statistics"]["silver"],
        "bronzeCount": item["statistics"]["bronze"]
    })
    
print(json.dumps(finalData, indent=4))
```
